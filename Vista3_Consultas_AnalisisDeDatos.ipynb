```python
#Seleccione tipo 'Departamento'
sel_departamento = datos_vivienda['Tipo'] == 'Departamento'
n1 = datos_vivienda[sel_departamento].shape[0]
n1
```

```python
#Seleccione tipo 'Casa y Habitacion'
seleccion = (datos_vivienda['Tipo'] == 'Casa') | (datos_vivienda ['Tipo'] == 'Habitación')
n2 = datos_vivienda[seleccion].shape[0]
n2
```

```python
#Seleccione tipo 'Area'
sel_areas = (datos_vivienda['Area'] >= 60 ) & (datos_vivienda['Area'] <= 100)
n3 = datos_vivienda[sel_areas].shape[0]
n3
```

```python
#Cuarto 4 y alquiler < 2000
selec_cuartos_alquiler = (datos_vivienda['Cuartos'] >= 4) & (datos_vivienda['Valor'] <= 2000)
n4 = datos_vivienda[selec_cuartos_alquiler].shape[0]
n4
```

```python
print(f"<-- El numero de departamentos es de: {n1}")
print(f"<-- El numero de Casa y habitaciones es de: {n2}")
print(f"<-- El numero de viviendas con un area de 60 y 100 m2 es de: {n3}")
print(f"<-- El numero de viviendas con al menos 4 habitaciones y cuestan menos de 2000 es de: {n4}")
```
