```python
barrios = ['Ate', 'Barranco', ' Comas', 'Lince', 'El Agustino', 'El agutino', 'San Luis', 'Callao']

seleccion = datos['Distrito'].isin(barrios)

datos = datos[seleccion]
```

```python
datos['Distrito'].drop_duplicates()
```

```python
grupo_barrio = datos.groupby('Distrito')
```

```python
grupo_barrio.groups
```

```python
for barrio, data in grupo_barrio:
  print('{} -> {}'.format(barrio, data.Valor.mean()))
```

```python
grupo_barrio[['Valor', 'Mantenimiento', 'Impuesto']].mean().round(2)
```

```python
grupo_barrio['Valor'].describe().round(2)
```

```python
grupo_barrio['Valor'].aggregate(['min', 'max']).rename(columns = {'min' : 'minimo', 'max' : 'maximo' })
```

# Visualizacion de datos

```python
%matplotlib inline

import matplotlib.pyplot as plt

plt.rc('figure', figsize = (20, 10))
```

```python
fig = grupo_barrio['Valor'].mean().plot.bar(color = 'gray')

fig.set_ylabel('Valor alquiler')

fig.set_title('Valor promedio alquiler por distrito', {'fontsize' : 22})
```

```python
fig = grupo_barrio['Valor'].max().plot.line(color = 'gray')

fig.set_ylabel('Valor alquiler')

fig.set_title('Valor maximo por distrito', {'fontsize' : 22})
```
